```python
df["Year"] = df["Date"].dt.year
df['Month'] = df["Date"].dt.month
df["Day"] = df["Date"].dt.day
df["DayOfYear"] = df["Date"].dt.dayofyear
df["DayOfWeek"] = df["Date"].dt.dayofweek
```

```python
from sklearn.preprocessing import MinMaxScaler
series = df["Close"].to_numpy().reshape(-1, 1)
```

```python
def create_sequence(data, seq_length):
  sequences = []
  labels = []

  for i in range(len(data) - seq_length):
    seq = data[i: i + seq_length]
    label = data[i + seq_length]
    sequences.append(seq)
    labels.append(label)
  return np.array(sequences), np.array(labels)
sequence_length = 2
X, y = create_sequence(series, sequence_length)
```

```python
from sklearn.model_selection import train_test_split
train_size = 0.7
X_train, y_train = X[:int(X.shape[0] * train_size)], y[:int(X.shape[0] * train_size)]
X_test, y_test = X[int(X.shape[0] * train_size):], y[int(X.shape[0] * train_size):]
```

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.LSTM(50, activation = "relu", input_shape = (sequence_length, 1)),
    tf.keras.layers.Dense(1)
])
model.compile(optimizer = tf.keras.optimizers.Adam(0.01), loss = "mse", metrics = [tf.keras.metrics.R2Score()])
```

```python
r = model.fit(X_train, y_train, epochs = 50, validation_data = (X_test, y_test), batch_size = 32)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
y_preds = model.predict(X_test)

mse = np.sqrt(mean_squared_error(y_test, y_preds))
mse
```

```python
forecast = pd.DataFrame({
    'Actual': y_test.reshape(-1),
    "Prediction": y_preds.reshape(-1)

})
plt.figure(figsize = (15, 8))
sns.lineplot(forecast["Actual"][:500], alpha = 0.5)
sns.lineplot(forecast["Prediction"][:500], alpha = 0.8)
```
