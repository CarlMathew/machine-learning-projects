```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, RocCurveDisplay, ConfusionMatrixDisplay
from sklearn.svm import SVC
from sklearn.ensemble import VotingClassifier

X = df.drop("diagnosis", axis = 1)
y = df["diagnosis"]

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.3, random_state = 32)

scaler = StandardScaler()
X_train, X_test = scaler.fit_transform(X_train), scaler.transform(X_test)

rfc = RandomForestClassifier(class_weight="balanced")
rfc.fit(X_train, y_train)

y_preds1 = rfc.predict(X_test)
print("RandomForestRegression")
print(classification_report(y_test, y_preds1))

fig, axes = plt.subplots(3,2, figsize = (15, 8))
RocCurveDisplay.from_predictions(y_test, y_preds1, ax= axes[0,0])
ConfusionMatrixDisplay.from_predictions(y_test, y_preds1, ax = axes[0,1])


svc = SVC(kernel = "linear", class_weight= "balanced")
svc.fit(X_train, y_train)

y_preds2 = svc.predict(X_test)
print("SVC")
print(classification_report(y_test, y_preds2))
RocCurveDisplay.from_predictions(y_test, y_preds2, ax= axes[1, 0])
ConfusionMatrixDisplay.from_predictions(y_test, y_preds2, ax = axes[1,1])

voting = VotingClassifier(estimators=[("rfc", rfc), ("svc", svc)])
voting.fit(X_train, y_train)
y_preds3 = voting.predict(X_test)
print("Voting")
print(classification_report(y_test, y_preds3))
RocCurveDisplay.from_predictions(y_test, y_preds3, ax= axes[2, 0])
ConfusionMatrixDisplay.from_predictions(y_test, y_preds3, ax = axes[2,1])
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import TensorDataset, DataLoader
```

```python
X = df.drop("diagnosis", axis = 1)
y = torch.tensor(df["diagnosis"].values).float()

scaler = StandardScaler()

transformed_X = torch.tensor(scaler.fit_transform(X)).float()

X_train, X_test, y_train, y_test = train_test_split(transformed_X, y, test_size= 0.3, random_state = 42)

train_data = TensorDataset(X_train, y_train)
train_loader =  DataLoader(train_data, shuffle = True, batch_size = 16, drop_last = True)

test_data = TensorDataset(X_test, y_test)
test_loader = DataLoader(test_data, shuffle = True, batch_size = test_data.tensors[0].shape[0], drop_last = True)
```

```python
class breast_classifier(nn.Module):
  def __init__(self):
    super().__init__()
    self.input = nn.Linear(30, 15)
    self.hidden1 = nn.Linear(15, 25)
    self.output = nn.Linear(25, 1)
  def forward(self, x):
    x = F.relu(self.input(x))
    x = F.relu(self.hidden1(x))
    x = self.output(x)
    return x
model = breast_classifier()
optim = torch.optim.Adam(model.parameters(), lr = 0.1)
lossfun = nn.BCEWithLogitsLoss()
```

```python
epochs = 500
losses_train = torch.zeros(epochs)
losses_test = torch.zeros(epochs)

acc_train = []
acc_test = []
for epoch in range(epochs):
  batchLoss = []
  model.train()
  batchAcc = []
  for X_train, y_train in train_loader:
    yHat = model(X_train)
    loss = lossfun(y_train, yHat.reshape(-1))

    optim.zero_grad()
    loss.backward()

    optim.step()
    yHat = [1 if i > 0 else 0 for i in yHat]
    acc = np.array(np.array(yHat) == np.array(y_train), dtype = int).mean()
    batchAcc.append(acc)
    batchLoss.append(loss.item())
  losses_train[epoch] = np.mean(batchLoss)
  acc_train.append(np.mean(batchAcc))

  model.eval()
  X_test, y_test =next(iter(test_loader))
  model.eval()
  with torch.no_grad():
    y_preds = model(X_test)
    loss_test = lossfun(y_test, y_preds.reshape(-1))
    losses_test[epoch] = loss_test
    y_preds = [1 if i > 0 else 0 for i in y_preds]
    acc_t = np.array(np.array(y_preds) == np.array(y_test), dtype = int).mean()
    acc_test.append(acc_t)
```
